```python
import pandas as pd
import numpy as np
from statsforecast import StatsForecast
from statsforecast.models import ARCH, GARCH, Naive
import yfinance as yf

from lib.statsforecast import preprocess_yf
from lib.utils import log_returns
```

```python
data_path = "../data/tests/AAPL_MSFT_2018-01-01_2022-12-13_1d.pkl"
df = pd.read_pickle(data_path)
```

```python
ppd = preprocess_yf(df, log_returns)
```

```python
models = [
    Naive(),
    ARCH(),
    GARCH()
]
```

```python
sf = StatsForecast(
    df = ppd, 
    models = models, 
    freq = "MS", # daily
    n_jobs = -1 # -1 selects all available cores
)
```

```python
levels = [80, 95] # confidence levels for the prediction intervals 
forecasts = sf.forecast(h=3, level=levels)
forecasts.reset_index(inplace=True)
forecasts.head()
```
