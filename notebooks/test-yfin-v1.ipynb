```python
import yfinance as yf
import pandas as pd 
```

```python
msft = yf.Ticker("MSFT")
```

```python
msft.options
```

```python
calls = msft.option_chain(msft.options[0]).calls
puts = msft.option_chain(msft.options[0]).puts
```

```python
df = pd.concat([calls, puts], axis=0)
```
